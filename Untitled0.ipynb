```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
frecuencia = 1  # Frecuencia de la señal en Hz
muestras = 1000  # Número de muestras
tiempo_total = 1  # Duración total de la señal en segundos
t = np.linspace(0.0, tiempo_total, muestras, endpoint=False)  # Vector de tiempo

# Señal cuadrada
senal_cuadrada = np.where(np.sin(2 * np.pi * frecuencia * t) > 0, 1, 0)

# Cálculo de la Transformada de Fourier
transformada_cuadrada = np.fft.fft(senal_cuadrada)

# Frecuencias correspondientes a las componentes de la transformada
frecuencias = np.fft.fftfreq(muestras, 1 / muestras)

# Graficar las señales y sus transformadas de Fourier
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, senal_cuadrada, label='Señal Cuadrada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(frecuencias, np.abs(transformada_cuadrada), label='Transformada Cuadrada')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
frecuencia = 1  # Frecuencia de la señal en Hz
muestras = 1000  # Número de muestras
tiempo_total = 1  # Duración total de la señal en segundos
t = np.linspace(0.0, tiempo_total, muestras, endpoint=False)  # Vector de tiempo

# Señal triangular
senal_triangular = np.abs(np.mod(t - 0.5 / frecuencia, 1) - 0.5) * 2

# Cálculo de la Transformada de Fourier
transformada_triangular = np.fft.fft(senal_triangular)

# Frecuencias correspondientes a las componentes de la transformada
frecuencias = np.fft.fftfreq(muestras, 1 / muestras)

# Graficar las señales y sus transformadas de Fourier
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, senal_triangular, label='Señal Triangular')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(frecuencias, np.abs(transformada_triangular), label='Transformada Triangular')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.legend()

plt.tight_layout()
plt.show()
```
